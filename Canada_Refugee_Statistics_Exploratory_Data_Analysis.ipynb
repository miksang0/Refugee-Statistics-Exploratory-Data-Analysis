```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
data = pd.read_csv('/content/can-stats-2012-22.csv')
```

```python
data.info()
```

```python
data.groupby('Country-of-origin')['UNHCR-refugees'].sum().sort_values(ascending=False).head()
```

```python
toprefugees = data.groupby('Country-of-origin')['UNHCR-refugees'].sum().sort_values(ascending=False).head(10).reset_index(name = 'Total')
```

```python
plt.xticks(rotation = 45)
sns.barplot(toprefugees, x = 'Country-of-origin', y = 'Total')
```
